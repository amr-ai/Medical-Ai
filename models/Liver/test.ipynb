```python
# Load model and predict on a new patient
loaded_model   = joblib.load('liver_disease_best_model.pkl')
loaded_scaler  = joblib.load('robust_scaler.pkl')
feature_names  = joblib.load('feature_names.pkl')

# New patient data (using exact feature names)
new_patient = pd.DataFrame([{
    'Age of the patient':              55,
    'Gender of the patient':           1,       # 1=Male, 0=Female
    'Total Bilirubin':                 3.0,
    'Alkphos Alkaline Phosphotase':    500,
    'Sgpt Alamine Aminotransferase':   500,
    'Sgot Aspartate Aminotransferase': 500,
    'Total Protiens':                  500,
    'ALB Albumin':                     500

}])

# Ensure column order matches training
new_patient = new_patient[feature_names]

# Scale and predict
new_patient_scaled = loaded_scaler.transform(new_patient)
prediction  = loaded_model.predict(new_patient_scaled)[0]
probability = loaded_model.predict_proba(new_patient_scaled)[0]

print('=' * 45)
print('       NEW PATIENT PREDICTION RESULT')
print('=' * 45)
print(f'  Prediction:           {"Liver Disease" if prediction == 1 else "Healthy"}')
print(f'  Confidence (Disease): {probability[1]*100:.1f}%')
print(f'  Confidence (Healthy): {probability[0]*100:.1f}%')
print('=' * 45)
if prediction == 1:
    print('  ⚠️  Patient may have liver disease — consult a doctor.')
else:
    print('  ✅  Patient appears healthy.')
```
